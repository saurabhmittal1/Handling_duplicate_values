```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
# Importing Dataset
data = pd.read_csv("../input/steam-video-games/steam-200k.csv", header = None)
data.shape
```

```python
data.head()
```

```python
# Changing all Column names to have better look at Data

data.columns = ['user_Id', 'game_title', 'behavior_name', 'value', 'no']
```

```python
data.head()
```

```python
# Getting Value Count in no column

data['no'].value_counts()
```

```python
# Dropping it because all the values are 0 and Same

data.drop(['no'], axis=1, inplace=True)
```

```python
#Sorting all data according to user_Id and game_title

sort_data = data.sort_values(['user_Id', 'game_title'], axis=0, ascending=True)
sort_data.shape
```

```python
# All Duplicate Columns

duplicate_col = sort_data[sort_data.duplicated(keep=False)]
duplicate_col.head(10)
```

```python
duplicate_col.shape
```

```python
# All Duplicated columns

sort_data[sort_data.duplicated()]
```

707 Columns have Duplicate values 

```python
# Dropping all duplicate values and keeping First Value

df = data.drop_duplicates(keep='first')
df.shape
```

```python
print("Rows before removing the duplicate values is : {}  ".format(data.shape[0]))
print("Rows after removing the duplicate values is  : {}  ".format(df.shape[0]))
print("Rows that have duplicate values is           : {}  ".format(data.shape[0] - df.shape[0]))
```
